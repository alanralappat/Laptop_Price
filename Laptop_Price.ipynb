# 2. Data Cleaning

```python
# drop Unnamed column from the dataset bcz its same as the index
df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Fill missing values in "data set" with mode
list1=["Company",	"TypeName",	"Inches",	"ScreenResolution",	"Cpu",	"Ram",	"Memory",	"Gpu",	"OpSys",	"Weight",	"Price"]
for i in list1:
  x=df[i].mode()[0]
  df[i] = df[i].fillna(x)
```

```python
df.isna().sum()
```

```python
# Replace "?" with NaN in the "Inches" column
df["Inches"] = df["Inches"].replace("?", pd.NA)

# Drop rows where "Inches" is NaN
df = df.dropna(subset=["Inches"])

# Convert "Inches" to float
df["Inches"] = df["Inches"].astype(float)

# Remove "GB" from "Ram" values and convert to float
df["Ram"] = df["Ram"].str.replace("GB", "").astype("float")

# Remove "kg" from "Weight" values
df["Weight"] = df["Weight"].str.replace("kg", " ")

# Replace "?" with NaN in the "Weight" column
df["Weight"] = df["Weight"].replace("?", pd.NA)

# Drop rows where "Weight" is NaN
df = df.dropna(subset=["Weight"])

# Convert "Weight" to float
df["Weight"] = df["Weight"].astype(float)

# Convert "Cpu" to categorical data type
df["Cpu"] = df["Cpu"].astype("category")

# Convert "Gpu" to categorical data type
df["Gpu"] = df["Gpu"].astype("category")

# Convert "OpSys" to categorical data type
df["OpSys"] = df["OpSys"].astype("category")

# Convert "Company" to categorical data type
df["Company"] = df["Company"].astype("category")
```

# Outlier detection

```python
import numpy as np
from scipy import stats

# Loop through each numeric column in the DataFrame
for i in df1.select_dtypes(include=[np.number]).columns:
   #This line selects only the numeric columns from the DataFrame.

    z_scores = np.abs(stats.zscore(df1[i]))
    #This computes the z-scores for the column, allowing us to find outliers based on how far a value is from the mean in terms of standard deviations.


    outliers = df1[z_scores > 3]  # Identify outliers
    print(f'Number of outliers in {i}: {outliers.shape[0]}')  # Print the number of outliers
```

```python
df1.plot(kind='box',subplots=True,layout=(8,3),figsize=(10,20))#Plotting a boxplot for each column in 'df1' using pandas' built-in plot functionality
```

```python
# Select numerical columns
numerical_columns =["Inches", "ScreenResolution", "Ram", "Memory", "Gpu","Weight","Cpu"]


# Create a copy of the original DataFrame to preserve the original data
df1_cleaned = df1.copy()

# Identify and remove outliers using IQR
for col in numerical_columns:
    Q1 = df1_cleaned[col].quantile(0.25)  # 25th percentile (Q1)
    Q3 = df1_cleaned[col].quantile(0.75)  # 75th percentile (Q3)
    IQR = Q3 - Q1  # Interquartile Range

    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove rows where the column values are outside the bounds
    df1_cleaned = df1_cleaned[(df1_cleaned[col]>= lower_bound) & (df1_cleaned[col] <= upper_bound)]

# Display the cleaned dataset
print("Data after removing outliers:")
print(df1_cleaned)
```

# Dimensionality Reduction

```python
#reduces the number of dimensions in a dataset while retaining most of the information
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the features before PCA
features =["Inches", "ScreenResolution", "Ram", "Memory", "Gpu","Weight", "Price","Cpu","OpSys"]
X = StandardScaler().fit_transform(df1_cleaned[features])

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Add PCA results to the DataFrame
df1_cleaned['pca1'] = pca_result[:, 0]
df1_cleaned['pca2'] = pca_result[:, 1]

# Visualize PCA results
sns.scatterplot(x='pca1', y='pca2', hue='Price', data=df1_cleaned)
plt.title('PCA of Price')
plt.show()
```

# Data Scalling

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Example DataFrame (replace this with your actual data)
# df = pd.read_csv('your_data.csv')  # Uncomment this if you're loading data from a CSV file

# Specify the columns to scale
columns_to_scale = ['Inches', 'ScreenResolution', 'Cpu', 'Ram', 'Memory', 'Gpu', 'Weight', 'Price']

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply MinMaxScaler to the specified columns
df1_cleaned[columns_to_scale] = scaler.fit_transform(df1_cleaned[columns_to_scale])

# Print the scaled DataFrame to verify
print("\nScaled DataFrame:\n")
print(df1_cleaned.head())

# Check statistics of the scaled data
print("\nStatistics of Scaled Data:\n")
print(df1_cleaned[columns_to_scale].describe())
```

## INITIAL MODELING AND HYPOTHESIS TEST

```python
# # Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# # Separate features (X) and target variable (y)
X = df1_cleaned.drop(columns=['Price'])  # All columns except the target
y = df1_cleaned['Price']  # Target column

# # Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=78)

# # Initialize the RandomForestRegressor model
model = RandomForestRegressor()

 # Train the model on the training data
model.fit(X_train, y_train)

# # Make predictions on the test data
y_pred = model.predict(X_test)

# # Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# # Print evaluation metrics
print("\nInitial Model Performance:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Accuracy (R^2 Score): {r2}")
```
